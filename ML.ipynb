```python
import pandas as pd
import sklearn.metrics as mt
```

```python
base = pd.read_csv('heart.csv')
```

```python
base.info()
```

```python
base.head()
```

```python
base.describe()
```

```python
from sklearn.model_selection import train_test_split

x = base.drop('target', axis=1)
y = base['target']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# logistic regressor
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()
log.fit(x_train, y_train)
```

```python
# test log
log_pred = log.predict(x_test)
print(mt.confusion_matrix(y_test, log_pred))
print(mt.accuracy_score(y_test, log_pred))
print(mt.f1_score(y_test, log_pred))
print(mt.precision_score(y_test, log_pred))
print(mt.recall_score(y_test, log_pred))
```

```python
# train log
log_pred_train = log.predict(x_train)
print(mt.confusion_matrix(y_train, log_pred_train))
print(mt.accuracy_score(y_train, log_pred_train))
print(mt.f1_score(y_train, log_pred_train))
print(mt.precision_score(y_train, log_pred_train))
print(mt.recall_score(y_train, log_pred_train))
```

```python
# random forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
# test rf
rf_pred = rf.predict(x_test)
print(mt.confusion_matrix(y_test, rf_pred))
print(mt.accuracy_score(y_test, rf_pred))
print(mt.f1_score(y_test, rf_pred))
print(mt.precision_score(y_test, rf_pred))
print(mt.recall_score(y_test, rf_pred))
```

```python
# train rf
rf_pred_train = rf.predict(x_train)
print(mt.confusion_matrix(y_train, rf_pred_train))
print(mt.accuracy_score(y_train, rf_pred_train))
print(mt.f1_score(y_train, rf_pred_train))
print(mt.precision_score(y_train, rf_pred_train))
print(mt.recall_score(y_train, rf_pred_train))
```

```python
# XGB
from xgboost import XGBClassifier

xgb = XGBClassifier()
xgb.fit(x_train, y_train)
```

```python
# test xgb
xgb_pred = xgb.predict(x_test)
print(mt.confusion_matrix(y_test, xgb_pred))
print(mt.accuracy_score(y_test, xgb_pred))
print(mt.f1_score(y_test, xgb_pred))
print(mt.precision_score(y_test, xgb_pred))
print(mt.recall_score(y_test, xgb_pred))
```

```python
# train xgb
xgb_pred_train = xgb.predict(x_train)
print(mt.confusion_matrix(y_train, xgb_pred_train))
print(mt.accuracy_score(y_train, xgb_pred_train))
print(mt.f1_score(y_train, xgb_pred_train))
print(mt.precision_score(y_train, xgb_pred_train))
print(mt.recall_score(y_train, xgb_pred_train))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import joblib

numeric_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
categorical_features = ['sex', 'fbs', 'exang', 'cp', 'restecg', 'slope', 'thal']


preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numeric_features),
    ('cat', OneHotEncoder(drop=None), categorical_features)
])


model = RandomForestClassifier()

pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('classifier', model)
])


pipeline.fit(x_train, y_train)


from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score

y_pred = pipeline.predict(x_test)
print(confusion_matrix(y_test, y_pred))
print("Acurácia:", accuracy_score(y_test, y_pred))
print("F1-Score:", f1_score(y_test, y_pred))
print("Precisão:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))


joblib.dump(pipeline, 'heart_pipeline.pkl')
```
